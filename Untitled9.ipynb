```python
# Identify numerical columns
numerical_cols = data.select_dtypes(include=['number']).columns

# Iterate through numerical columns and remove outliers using IQR
for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
```

```python
#feature and Target
X = data.drop('Heart_Disease', axis=1)
y = data['Heart_Disease']
```

```python
import matplotlib.pyplot as plt
y.value_counts().plot(kind='bar')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.title('Distribution of Heart Disease')
plt.show()
```

```python
# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Assuming 'General_Health' is the only categorical column in X
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode categorical features
label_encoder = LabelEncoder()

# Identify all categorical columns, including 'General_Health'
categorical_cols = X.select_dtypes(include=['object']).columns

# Encode all categorical columns
for col in categorical_cols:
    X[col + '_Encoded'] = label_encoder.fit_transform(X[col])
    X = X.drop(col, axis=1)  # Remove original columns after encoding

# Now normalize the numerical features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y_encoded, test_size=0.3, random_state=42)

# Import the necessary module
from keras.utils import to_categorical

# Convert target to categorical (one-hot encoding)
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
# Build the ANN model
model = Sequential()
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(2, activation='softmax'))  # Assuming binary classification
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train_categorical, validation_data=(X_test, y_test_categorical), epochs=50, batch_size=32)
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test_categorical)
print(f'Test accuracy: {test_accuracy:.2f}')
```

```python
# Plotting accuracy and loss over epochs
# Accuracy plot
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# function to create different models
def create_model(layers, activation, optimizer='adam'):
    model = Sequential()
    for i, neurons in enumerate(layers):
        if i == 0:
            model.add(Dense(neurons, input_dim=X_train.shape[1], activation=activation))
        else:
            model.add(Dense(neurons, activation=activation))
    model.add(Dense(2, activation='softmax'))  # Assuming binary classification
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# List of configurations to try
configurations = [
    {'layers': [32, 16], 'activation': 'relu', 'optimizer': 'adam'},
    {'layers': [64, 32, 16], 'activation': 'relu', 'optimizer': 'adam'},
    {'layers': [32, 16], 'activation': 'sigmoid', 'optimizer': 'adam'},
    {'layers': [32, 16], 'activation': 'relu', 'optimizer': 'sgd'},
    {'layers': [32, 16], 'activation': 'tanh', 'optimizer': 'rmsprop'},
]

# Store results
results = []
```

```python
# Train and evaluate each configuration
for i, config in enumerate(configurations):
    print(f"Training model {i+1} with configuration: {config}")
    model = create_model(config['layers'], config['activation'], config['optimizer'])
    history = model.fit(X_train, y_train_categorical, validation_data=(X_test, y_test_categorical), epochs=50, batch_size=32, verbose=0)
    test_loss, test_accuracy = model.evaluate(X_test, y_test_categorical, verbose=0)
    results.append({
        'config': config,
        'test_accuracy': test_accuracy,
        'history': history
    })
    print(f"Test accuracy: {test_accuracy:.4f}")
```

```python
# Plotting accuracy and loss for each configuration
for i, result in enumerate(results):
    plt.plot(result['history'].history['val_accuracy'], label=f"Config {i+1} - Val Accuracy")

plt.title('Validation Accuracy for Different Configurations')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

```

```python
for i, result in enumerate(results):
    plt.plot(result['history'].history['val_loss'], label=f"Config {i+1} - Val Loss")

plt.title('Validation Loss for Different Configurations')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
